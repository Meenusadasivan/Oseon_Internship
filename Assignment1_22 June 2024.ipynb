# Task 1

```python

import numpy as np
# Set the random seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 1000

# Generate Age:  mean 35 and standard deviation 10
age = np.random.normal(35, 10, n_samples)

# Generate Height: mean 170 cm and standard deviation 15 cm
height = np.random.normal(170, 15, n_samples)

# Generate Weight: mean 70 kg and standard deviation 10 kg
weight = np.random.normal(70, 10, n_samples)

# Generate Gender: Randomly assigned with 50% probability for 'Male' and 'Female'
gender = np.random.choice(['Male', 'Female'], n_samples)

# Generate Income:mean 50,000 and standard deviation 15,000
income = np.random.normal(50000, 15000, n_samples)

# Combine all the features into a single array
dataset = np.column_stack((age, height, weight, gender, income))

# Print the first 10 samples of the dataset
print("First 100 samples of the dataset:")
print(dataset[:100])
```

```python
import pandas as pd

```

```python
# Create a pandas DataFrame
data = pd.DataFrame({
    'Age': age,
    'Height': height,
    'Weight': weight,
    'Gender': gender,
    'Income': income
})

```

```python
# Display the first few rows of the dataset
print(data.head())
```

## Task 2- Descriptive Statistics

#### Task 2-mean, median, standard deviation, and variance for Age, Height, Weight, and Income.

```python
# Calculate mean, median, standard deviation, and variance
mean_values = data[['Age', 'Height', 'Weight', 'Income']].mean()
median_values = data[['Age', 'Height', 'Weight', 'Income']].median()
std_values = data[['Age', 'Height', 'Weight', 'Income']].std()
var_values = data[['Age', 'Height', 'Weight', 'Income']].var()
```

```python
print("Mean value=",mean_values)
print("Median Value=",median_values)
print("Standard Deviation=",std_values)
print("Variance=",var_values)
```

## Task 3: Data Visualization

#### 4. Plot histograms for Age, Height, Weight, and Income

```python
# import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set(style="whitegrid")

# Plot histograms for Age, Height, Weight, and Income
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Age
sns.histplot(data['Age'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')

# Height
sns.histplot(data['Height'], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Height Distribution')

# Weight
sns.histplot(data['Weight'], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Weight Distribution')

# Income
sns.histplot(data['Income'], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Income Distribution')

plt.show()
```

#### Use Seaborn to create KDE plots for Age, Height, Weight, and Income. 

```python
sns.set(style="whitegrid")

# Create KDE plots for Age, Height, Weight, and Income
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Age
sns.kdeplot(data['Age'], shade=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')

# Height
sns.kdeplot(data['Height'], shade=True, ax=axes[0, 1])
axes[0, 1].set_title('Height Distribution')

# Weight
sns.kdeplot(data['Weight'], shade=True, ax=axes[1, 0])
axes[1, 0].set_title('Weight Distribution')

# Income
sns.kdeplot(data['Income'], shade=True, ax=axes[1, 1])
axes[1, 1].set_title('Income Distribution')


plt.show()
```

#### 5.boxplots for Age, Height, Weight, and Income

## Task 4: Correlation Analysis

####  6.Pearson correlation coefficient between Age, Height, Weight, and Income. 

```python
corr_matrix = data[['Age', 'Height', 'Weight', 'Income']].corr()

print(corr_matrix)
```

#### 7. Heatmap to visualize the correlation matrix. 


```python
# Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix , annot=True, cmap='coolwarm')
plt.title("Correlation matrix")
plt.show()
```

## Task 5: Inferential Statistics 

#### 8. Hypothesis test-test to see if there is a significant difference in Income between Male and Female

```python
# H0: there is no significant difference in Income between Male and Female
# H1: there is a significant difference in Income between Male and Female
import scipy.stats as stats
# Separate the income data based on gender
income_male = data[data['Gender'] == 'Male']['Income']
income_female = data[data['Gender'] == 'Female']['Income']

# Perform an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(income_male, income_female)
print(t_stat, p_value)
```
